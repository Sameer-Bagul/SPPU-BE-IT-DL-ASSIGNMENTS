```python
import numpy as np
data= np.array([[0,   0,   1/2, 0],
               [1/3, 0,   0,   1/2],
               [1/3, 1/2, 0,   1/2],
               [1/3, 1/2, 1/2, 0]])

n=data.shape[0]
d=0.85
tollerence=0.001
r=np.ones(n)/n
print(r)

while True:
     new_r=(1-d)/n+d*data.dot(r)
     if(np.linalg.norm(new_r-r,ord=1)<tollerence):
         break
     r=new_r

print("Hii")
```
